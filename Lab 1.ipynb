```python
import random
import matplotlib.pyplot as plt

twoDigitCollision = []

for i in range(200):
    
    notFound = True

    # Make a new array of size 100 and fill it with 0
    digitList = [0]*100

    # Keep track of how many runs until a collision
    count = 0

    # While we have no collision keep trying
    while notFound:
        studentId = random.randint(999999,10000000)

        count += 1

        # Extract the last 2 digits of the studentId
        lastTwo = studentId % 100

        # If the index has already been generated, we have a collision
        if digitList[lastTwo] == 1:
            notFound = False
        # Otherwise, indicate that this index has been generated
        else:
            digitList[lastTwo] = 1

    # Printing the number of attempts it took
    #print(count)
    
    # Adding the data to a list to plot it later
    twoDigitCollision.append(count)
    
# scatter plot

plt.scatter(range(200), twoDigitCollision)

# add title
plt.title("Scatter Plot")
 
# set X and Y labels
plt.xlabel('Trial')
plt.ylabel('Attempts before Collision')
 
plt.show()
```
